```python
# drop the rows that have nan values.
drop_na_cols = ["TAIL_NUM", "DEP_TIME", "DEP_DELAY", "ARR_TIME", "ARR_DELAY", "AIR_TIME"]
df = df.dropna(subset=list(drop_na_cols))
df.shape
```

**observation**

- dropped 4224 (1%) of the data due to incomplete information

- each flight is labelled with FLIGHTS = 1

- ID/ city names are categorical

- departure time and arrival time are time in string format


```python
# convert the time columns
df["DEP_TIME"] = df["DEP_TIME"].astype("int").astype("str")
df["ARR_TIME"] = df["ARR_TIME"].astype("int").astype("str")
```

```python
# cast arrival and depature time columns 
def cast_time_columns(x):
    x = x.zfill(4)
    x = '0000' if x=='2400' else x
    x = x[0:2] + ":" + x[2:4]
    return x

df["DEP_TIME"] = df["DEP_TIME"].apply(lambda x: cast_time_columns(x))
df["ARR_TIME"] = df["ARR_TIME"].apply(lambda x: cast_time_columns(x))

df.head()
```

## 8. predictive analytics

as the most relavant feature with arraval delay is depature delay, we can mostly certain to predict using only one feature

```python
# Import the model we are using
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
df = df.copy()
df["weekday"] = pd.to_datetime(df["FL_DATE"]).dt.weekday
```

```python
# Labels are the values we want to predict
labels = np.array(df["ARR_DELAY"])

# select feature columns
features_cols = df[["DEP_DELAY", "OP_CARRIER_AIRLINE_ID", "ORIGIN_AIRPORT_ID", "DEST_AIRPORT_ID", "AIR_TIME", "weekday"]]

features_cnames = features_cols.columns

# convert to numpy array
features = np.array(features_cols)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.20, random_state=42)
```

```python
# Instantiate model ( default with 100 decision trees)
base_model = RandomForestRegressor(random_state = 42)

# Train the model on training data
base_model.fit(X_train, y_train)

# Use the forest's predict method on the test data
predictions = base_model.predict(X_test)
```

```python
def evaluate(model, test_features, test_labels):
    """ print out the model evaluation result. metric used are accuracy and MAE """
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mae = round(np.mean(errors), 2)
    
    r2 = r2_score(test_labels, predictions)
    print('---------------Model Performance---------------')
    print(f'Mean Absolute Error: {mae}.')
    print(f'R2: {r2}.')
    print('-----------------------------------------------')
```

```python
evaluate(base_model, X_test, y_test)
```

### predicted arrival delay vs real arrival delay

```python
plt.figure(figsize=(20,5))
matplotlib.rcParams.update({'font.size': 15})

plt.subplot(121)
plt.plot(y_test, predictions, '.')
plt.xlabel("real arrival delay")
plt.ylabel("predicted arrival delay")


plt.subplot(122)
plt.plot(y_test, predictions-y_test, '.')
plt.xlabel("real arrival delay")
plt.ylabel("prediction delta")
plt.show()
```

### Feature importance understanding

```python
plt.figure(figsize=(15,7))
matplotlib.rcParams.update({'font.size': 15})

plt.barh(features_cnames, base_model.feature_importances_)
plt.xlabel("feature importance")
plt.ylabel("feature names")

plt.show()
```
